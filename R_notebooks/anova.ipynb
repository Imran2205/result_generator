```r
options(scipen=999)
dir_files <- "/Users/imrankabir/Desktop/research/vqa_accessibility/result_generator/data/sim_data_fina/csv_long/"
```

```r
for (file in list.files(dir_files)){
    print('#############################################################')
    print('#############################################################')
    print(file)
    data <- read.csv(paste(dir_files, file, sep=""))
    res.aov <- aov(Value ~ Label, data = data)
    print(summary(res.aov))
    val <- data$Value
    lbl <- data$Label
    print('*****************************')
    print(pairwise.t.test(val, lbl, p.adj = "bonf"))
}
```
